# Visualization of Sampling Distribution

```python
# Loads packages
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mtcars = pd.read_csv("https://data-analytics.zybooks.com/mtcars.csv")
```

```python
# Prints the first few lines of mtcars
print(mtcars.head())
```

```python
# Show summary statistics for the variable wt
mtcars.wt.describe()
```

## Boxplots

An axis with the minimum and maximum data points as the endpoints shows the range of the data.
  1. The median is the middle number in the ordered data and is represented as a line inside the box.
  2. **Q1** is the median of the lower half of the data and forms the lower boundary of the box.
  3. **Q3** is the median of the upper half of the data and forms the upper boundary of the box. 

The box is formed and the whiskers are extended from the lower boundary of the box to the minimum and from the upper boundary of the box to the maximum.

```python
# Plots a box plot
sns.boxplot(data=mtcars.wt, width=0.35)
```

## Violin Plots

A violin plot displays the distribution of the data in a dataset in ways similar to a box plot, except that a violin plot includes information on the relative number of data points with any given value. A violin plot gives information on the median, interquartile range, the interval in which the true value of the population mean lies  of the time, and the shape of the data.

```python
# Load the attention dataset
att = sns.load_dataset("attention")
```

```python
# Make plot
ax = sns.violinplot(x="score", data=att)

# Show the image
plt.show()
```

```python
# Make a side-by-side violin plot
ax = sns.violinplot(x="solutions", y="score", data=att)

plt.show()
```

```python
# Make a side-by-side violin plot
ax = sns.violinplot(x="solutions", y="score", data=att)

plt.show()
```

```python
# Add another variable to the plot
ax = sns.violinplot(x="solutions", y="score",
                    hue="attention", data=att, palette="muted")
ax.legend(loc="lower right")
plt.show()
```
