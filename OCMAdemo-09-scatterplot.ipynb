## Scatter Plots

To graph a scatter plot, the `sns.regplot(x,y, ci = None)` function of the seaborn library is used, which takes in two arrays of equal size x and y. By default, the  confidence intervals are displayed, but the parameter ci can be set to None to disable confidence intervals.
The regplot can also be used with data frames. Ex: Using the rock dataset, rock area is plotted with respect to rock perimeter using the second block of code below.

```python
# Load the necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# Load the rock data set
df = pd.read_csv("https://static-resources.zybooks.com/DataScience/rock.csv")
```

```python
# Set title
plt.title('Area and perimeter of rock', fontsize=20)

# define plot
sns.regplot(x=df["Area"], y=df["Perimeter"], ci=None, fit_reg=False)

# show the image
plt.show()
```
