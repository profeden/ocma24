# Pair Plots

A pairplot displays pairwise relationships between a datasets variables and plots all numeric variables by default. Its essentially a grid of subplots, where each subplot displays a scatterplot of a pair of variables from the dataset, along with the corresponding histograms or kernel density estimates (KDEs) of each variable along the diagonal. Pairplots with Matplotlib usually involve manual iteration over the dimensions of the data. For each pair of variables in a dataset, seaborn offers a useful `pairplot()` function that allows us to create pairplots using scatterplots and histograms.


```python
sns.set_theme()    # set default theme
peng_df = sns.load_dataset('penguins')

sns.pairplot(data = peng_df, dropna = True)
```
